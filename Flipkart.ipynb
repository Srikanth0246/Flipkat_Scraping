```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3DRealme&otracker=nmenu_sub_Electronics_0_Realme'
response = requests.get(url)
print(response.content)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
products = soup.find_all('div', {'class': '_2kHMtA'})
```

```python
for i in products:
  print(i,"----\n\n")
```

```python
title=[]
price=[]
for product in products:
    title.append(product.find('div',{'class': '_4rR01T'}).text)
    price.append(product.find('div', {'class': '_30jeq3 _1_WHN1'}).text)
```

```python
for i in range(len(title)):
  print(title[i]," ",price[i])
```
